```python
import gymnasium as gym  # Note: we use gymnasium instead of gym
from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy
from gymnasium.wrappers import RecordVideo
import os
```

```python
os.makedirs("videos/training", exist_ok=True)
os.makedirs("videos/testing", exist_ok=True)
os.makedirs("videos/running", exist_ok=True)
```

```python
train_env = gym.make("LunarLander-v3", render_mode="rgb_array")
train_env = RecordVideo(train_env, video_folder="videos/training", episode_trigger=lambda ep: True)
```

```python
model = PPO(
    "MlpPolicy",
    train_env,
    verbose=1,
    learning_rate=3e-4,
    gamma=0.99,
    batch_size=128,
    n_steps=512,
    ent_coef=0.02,
    clip_range=0.1,
    n_epochs=10
)
```

```python
model.learn(total_timesteps=500_000)
model.save("ppo_lunarlander")
train_env.close()
```

```python
# Load the saved model
model = PPO.load("ppo_lunarlander")

# Create a testing environment
test_env = gym.make("LunarLander-v3", render_mode="rgb_array")
test_env = RecordVideo(test_env, video_folder="videos/testing", episode_trigger=lambda ep: True)

# Evaluate the model
mean_reward, std_reward = evaluate_policy(model, test_env, n_eval_episodes=10)
print(f"Mean reward: {mean_reward} +/- {std_reward}")
test_env.close()
```
